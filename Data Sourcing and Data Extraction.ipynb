<font size= "3" color="orange"><b><i>a. Scrape top played track (first only) on SoundCloud

```python
# specify the url
link = "https://soundcloud.com/charts/top"

# query the website and return the html to the 'page' variable
page = urllib2.urlopen(link)
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page, 'html.parser')
print soup.prettify()
```

```python
# indetify the correct table
table_topTracks = soup.find('section', class_= 'sounds')
```

```python
# identify the top track
top_track = table_topTracks.find('article')
top_track 
```

```python
# get relevant information
track = top_track.find_all('a')
title = track[0].find(text=True)
artist = track[1].find(text=True)

# print result
print ('The latest top played track on SoundCloud is ' + title + ' by ' + artist + '.')
```

<font size= "3" color="orange"><b><i> b. Scrape table for the Cinemalaya Philippine Independent Film Festival


```python
# libararies used: urllib2, pandas, BeautifulSoup

# specify the url
wiki = "https://en.wikipedia.org/wiki/Cinemalaya_Philippine_Independent_Film_Festival"

# query the website and return the html to the 'page' variable
wiki_page = urllib2.urlopen(wiki)

#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(wiki_page, 'html.parser')
print soup.prettify()
```

```python
# among the four tables available from the page, scrape New Breed Table
all_table = soup.find_all('table', class_="wikitable")
new_breed = all_table[0].find()
# director_showcase = all_table[1].find()
# main_competition = all_table[2].find()
# short_film = all_table[3].find()
```

```python
# generate empty list
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

#fill each list row by row
for row in new_breed.find_all("tr"):
    print(row) # for debugging 
    cells = row.find_all('td') 
    film = row.find_all('b') # to store second column data
    if len(cells)>0: # Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(film[0].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
```

```python
df = pd.DataFrame(A,columns=['Year'])
df['Best Film']=B
df['Best Direction']=C
df['Special Jury Prize']=D
df['Best Actress']=E
df['Best Actor']=F
df
```
